```python
import matplotlib.pyplot as plt
import numpy as np

# Define the line
line_x = np.array([0, 1])
line_y = np.array([1, 0])

# Define the circle
theta = np.linspace(0, 2*np.pi, 100)
circle_x = np.cos(theta)
circle_y = np.sin(theta)

# Define the triangle
triangle_x = np.array([0, 0, 1, 0])
triangle_y = np.array([0, 1, 0, 0])

# Plot the area
plt.fill_between(triangle_x, triangle_y, color='lightgray', label='Area [0,3]x[0,3]')

# Plot the line
plt.plot(line_x, line_y, label='Line: (0,1) to (1,0)')

# Plot the circle
plt.plot(circle_x, circle_y, label='Circle', linestyle='dashed')

# Plot the triangle
plt.plot(triangle_x, triangle_y, label='Triangle', linestyle='dashed')

# Plot the cones
cone_theta = np.linspace(0, np.pi, 100)
cone_x = np.cos(cone_theta)
cone_y = np.sin(cone_theta)

plt.plot(cone_x, cone_y, linestyle='dashed', color='red', label='Cone 1')
plt.plot(-cone_x, -cone_y, linestyle='dashed', color='red', label='Cone 2')

plt.xlim([-1, 2])
plt.ylim([-1, 2])

plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Circle parameters
radius = 0.25
center = (0.5, 0.5)

# Equation of the circle
def circle_equation(x, center, radius):
    return np.sqrt(radius**2 - (x - center[0])**2) + center[1]

# Plot the circle
theta = np.linspace(0, 2*np.pi, 100)
x_circle = center[0] + radius * np.cos(theta)
y_circle = center[1] + radius * np.sin(theta)

plt.figure(figsize=(8, 8))

plt.plot(x_circle, y_circle, label='Circle')

import numpy as np
import matplotlib.pyplot as plt

# Function to calculate endpoint of the line given angle t
def calculate_endpoint(angle):
    x_endpoint = np.cos(angle)
    y_endpoint = np.sin(angle)
    return x_endpoint, y_endpoint

# Angle in radians
distance = np.sqrt(center[0]**2 + center[1]**2)
angle_t = np.arcsin(radius / distance)

print(radius, distance, angle_t)

# Calculate endpoint
x_endpoint, y_endpoint = calculate_endpoint(angle_t)

print(f'Angle t: {angle_t:.2f} degrees')
print(f'Endpoint: ({x_endpoint:.2f}, {y_endpoint:.2f})')

# Plot the line
plt.plot([0, x_endpoint], [0, y_endpoint], label=f'Line at {np.degrees(angle_t):.2f} degrees')
plt.plot([0, x_endpoint], [0, y_endpoint], label=f'Line at {np.degrees(angle_t):.2f} degrees')


# plt.plot(x, y_1, label='Tangent Line (Through Origin)')
# plt.plot(x, y_2, label='Tangent Line (Through Origin)')

# Set axis limits
plt.xlim(center[0] - radius - 0.1, center[0] + radius + 0.1)
plt.ylim(center[1] - radius - 0.1, center[1] + radius + 0.1)

# Set labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Circle with Tangent Line Through Origin')

plt.xlim([0, 2])
plt.ylim([0, 2])

# Show legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
